```python


risk_df = df[df['biopsed']==False].copy()

# Step 1a — grew, changed, bleed, hurt, elevation
# Yeh STRING hain — 'TRUE'/'FALSE' capital mein
cols_boolean = ['grew','changed','bleed','hurt','elevation']
risk_df[['score_grew','score_changed','score_bleed',
         'score_hurt','score_elevation']] = risk_df[cols_boolean].replace({
    'TRUE': 1,
    'FALSE': 0
})

# Step 1b — smoke aur cancer_history alag!
risk_df['score_smoke'] = risk_df['smoke'].replace({
    True: 1, False: 0, 'Unknown': 0.5
})
risk_df['score_cancer_history'] = risk_df['cancer_history'].replace({
    True: 1, False: 0, 'Unknown': 0.5
})

# Step 2 — Diameter
risk_df['score_diameter_1'] = (risk_df['diameter_1'] > 15).astype(int)
risk_df['score_diameter_2'] = (risk_df['diameter_2'] > 15).astype(int)

# Step 3 — Total score
risk_df['total_risk_score'] = risk_df[[
    'score_grew','score_changed','score_bleed',
    'score_hurt','score_elevation','score_smoke',
    'score_cancer_history','score_diameter_1','score_diameter_2'
]].sum(axis=1)

# Step 4 — Category
risk_df['risk_category'] = pd.cut(
    risk_df['total_risk_score'],
    #bins=[-0.1,3,6,9],
    bins=[-0.1, 2, 3.5, 5.1],
    labels=['Low Risk', 'Medium Risk', 'High Risk']
)

print(risk_df['risk_category'].value_counts())
```

```python
# Step 1 — Columns select karo aur score ke hisaab se sort karo
heatmap_df = risk_df[[
    'patient_id', 'diagnostic', 'age_group',
    'total_risk_score', 'risk_category',
    'grew', 'changed', 'bleed', 'hurt', 'elevation',
    'smoke', 'cancer_history', 'diameter_1', 'diameter_2'
]].sort_values('total_risk_score', ascending=False)

# Step 2 — Color function banao
def color_risk(val):
    if val == 'High Risk':
        return 'background-color: Red'
    elif val == 'Medium Risk':
        return 'background-color: Yellow'
    else:
        return 'background-color: Green'

# Step 3 — Styled object banao
styled = heatmap_df.style

# Step 4 — risk_category column mein color lagao
styled = styled.map(
    color_risk,
    subset=['risk_category']
)

# Step 5 — total_risk_score mein gradient color lagao
styled = styled.background_gradient(
    subset=['total_risk_score'],
    cmap='RdYlGn_r'
)

# Step 6 — Excel mein save karo
styled.to_excel(
    'skin_cancer_heatmap.xlsx',
    engine='openpyxl',
    index=False
)

print("Heatmap Excel saved!")

#Newer Pandas mein applymap ka naam badal gaya hai — map ho gaya!
```
